```python
r=3.
x0=0.1
steps=1000

def generate_data(r,x0=0.1,steps=1200):
    ''' 
    Write and read the last 200 values.
    '''
    with open('data.txt','w+') as file:   
        for t in range(steps):
            x=logistic_map(x0,r)
            file.write(str(x)+f'\n')
            x0=x
        file.seek(0)
        fileR = file.readlines()[-201:-1]
        return fileR

data = generate_data(r,x0)
data
```
